```python
#cleaning the text and having only alphanumeric characters
def cleanTxt(text):
    text = re.sub(r'[^\w]', ' ', text)
    return text


data['Comment'] = data['Comment'].apply(cleanTxt)
```

```python
#getting subjectivity and polarity to predict the sentiment of the comment
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# get polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity


#Columns
data['Subjectivity'] = data['Comment'].apply(getSubjectivity)
data['Polarity'] = data['Comment'].apply(getPolarity)

data
```

```python
#depending on polarity comment is categorized to be positive negative or neutral
def getAnalysis(score):
    if score < 0 :
        return 'Negative'
    elif score > 0:
        return 'Positive'
    else:
        return 'Neutral'
    
data['TextBlob Analysis'] = data['Polarity'].apply(getAnalysis)
```

```python
from transformers import pipeline
classifier = pipeline("text-classification")
```

```python
# using hugging face - pipeline to predict the sentiment
text = data["Comment"]
text
ans = []
for i in text:
    j = classifier(i)
    print(i,j[0]["label"])
    ans.append(j[0]["label"])
data['HuggingFace Analysis'] = ans
```

```python
data["TextBlob Analysis"]=data["TextBlob Analysis"].str.capitalize()
data["HuggingFace Analysis"] = data["HuggingFace Analysis"].str.capitalize()
```

```python
positive_count = (data['TextBlob Analysis'] == 'Positive').sum()
negative_count = (data['TextBlob Analysis'] == 'Negative').sum()
neutral_count = (data['TextBlob Analysis'] == 'Neutral').sum()

# Calculate the total number of comments
total_comments = len(data)

# Calculate the percentage of positive, negative, and neutral comments
positive_percentage = (positive_count / total_comments) * 100
negative_percentage = (negative_count / total_comments) * 100
neutral_percentage = (neutral_count / total_comments) * 100

print("TextBlob Analysis: ")
# Check if the positive comments are 50% more than the negative comments
if positive_percentage >= 1.5 * negative_percentage:
    print("The overall sentiment of users is positive.")
elif neutral_percentage > positive_percentage and neutral_percentage > negative_percentage:
    print("The overall sentiment of users is neutral.")
else:
    print("The overall sentiment of users is negative.")
```
